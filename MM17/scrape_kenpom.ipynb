```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

### Adjust url to loop between 2003 and 2017 and retrieve info using html attributes.

#### Save to dicts where each key is a year and every value is the pandas DF.

```python
info_list = []
dicts = {}
years = [2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
for year in years:
#try clause to skip any companies with missing/empty board member tables

#loop through table
    base_url = f'https://kenpom.com/index.php?y={year}'
    response = requests.get(base_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    data_table = soup.find('table', {"id" : "ratings-table"})
    for row in data_table.find_all('tbody'):
        for thing1 in row.find_all('tr'):
            cols = thing1.find_all('td')
            if len(cols) != 21:
                continue
            else:
                info_list.append(([cols[num].text.strip() for num in range(0,21)]))
    rankings = np.asarray(info_list)
    df = pd.DataFrame(rankings)
    #rename columns, check output
    df.columns = list(map(str, range(0,21)))
    good_cols = list(map(str, [0,1,2,3,4,5,7,9,11,13,15,17,19]))
    df = df[good_cols]
    columns = ['Rk','Team', 'Conf', 'W-L',  'AdjEM', 'AdjO ', 'AdjD' , 'AdjT', 'Luck', 'AdjEM', 'OppO', 'OppD', 'AdjEM']
    df.columns = columns
    print(str(year))
    if str(year) not in dicts.keys():
        dicts[str(year)] = df
        df.to_csv(f"{year}_sos_stats.csv")
        print(f'Dict {year} is done.')
```
